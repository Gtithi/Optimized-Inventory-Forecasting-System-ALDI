```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import matplotlib.pyplot as plt

# Example Data Preparation
# Assuming df_old is already your dataframe with 'inventory' column and 'date' as index
df_old.index = pd.to_datetime(df_old.index)
df_old.sort_index(inplace=True)  # Ensure the data is sorted by date

# Split Data into Train and Test sets
train_size = int(len(df_old) * 0.8)  # Use 80% of data for training
train, test = df_old.iloc[:train_size], df_old.iloc[train_size:]

# Fit the Holt-Winters Exponential Smoothing Model
model = ExponentialSmoothing(train['inventory'], trend = 'add',seasonal='add', seasonal_periods=130).fit(smoothing_level=0.7, smoothing_slope=0.5, smoothing_seasonal=0.2)

# Forecast for the Test Period
test_forecast = model.forecast(len(test))

# Forecast for the Next 7 Days
future_forecast = model.forecast(7)

# Plot the Results
plt.figure(figsize=(12, 6))
#plt.plot(train.index, train['inventory'], label='Train')
plt.plot(test.index, test['inventory'], label='Test')
plt.plot(test.index, test_forecast, label='Test Forecast')
plt.plot(pd.date_range(start=test.index[-1], periods=8, closed='right'), future_forecast, label='7-Day Forecast', color='red')
plt.legend(loc='best')
plt.title('Holt-Winters Exponential Smoothing Forecast')
plt.xlabel('Date')
plt.ylabel('Inventory')
plt.show()

# Print the forecasted values for the next 7 days
print("Forecast for the next 7 days:")
print(future_forecast)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from scikeras.wrappers import KerasRegressor
import matplotlib.pyplot as plt

# Assume df_new is your DataFrame with 'cal_date' as the index
# and columns: 'sales', 'inventory', 'delivery', 'demand_forecast'

# Add a lagged delivery feature (lag 1)
df_new['delivery_lag1'] = df_new['delivery'].shift(1)

# Drop any rows with NaN values due to the shift
df_new.dropna(inplace=True)

# Define your features (X) and target variable (y)
features = ['delivery_lag1']
target = 'inventory'

X = df_new[features].values
y = df_new[target].values.reshape(-1, 1)

# Scale the data using MinMaxScaler
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# Split the data into train and test sets (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.3, shuffle=False)

# Reshape the input data to be 3D for LSTM [samples, time steps, features]
X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

```

```python
# Function to create LSTM model
def create_lstm_model(units=50, optimizer='adam'):
    model = Sequential()
    model.add(LSTM(units, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
    model.add(Dense(1))
    model.compile(optimizer=optimizer, loss='mse')
    return model

# Wrap Keras model using KerasRegressor
keras_reg = KerasRegressor(build_fn=create_lstm_model, verbose=0)

# Define parameters for grid search
param_grid = {
    'units': [50,
    'optimizer': ['adam', 'rmsprop']
}

# Use TimeSeriesSplit for cross-validation (to preserve temporal order)
tscv = TimeSeriesSplit(n_splits=3)

# Grid search with cross-validation
grid_search = GridSearchCV(estimator=keras_reg, param_grid=param_grid, cv=tscv, scoring='neg_mean_squared_error', verbose=1)
grid_search.fit(X_train_reshaped, y_train)

# Best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Lowest RMSE found: ", np.sqrt(-grid_search.best_score_))

# Retrieve the best model from grid search
best_lstm_model = grid_search.best_estimator_.model

```

```python
# Number of days to forecast
forecast_horizon = 7

# Initialize the last known inventory value for prediction
last_known_inventory = y[-1]  # Use the last value from the original data (unscaled)

# List to store forecasted values
forecasted_values = []

# Iterate over the forecast horizon
for day in range(forecast_horizon):
    # Scale the last known inventory value
    last_known_inventory_scaled = scaler_y.transform(last_known_inventory.reshape(-1, 1))

    # Reshape for LSTM input [samples, time steps, features]
    X_pred = np.array([last_known_inventory_scaled]).reshape(1, 1, 1)

    # Predict the next inventory value
    y_pred_scaled = model.predict(X_pred)

    # Inverse transform to get the actual inventory value
    y_pred = scaler_y.inverse_transform(y_pred_scaled)

    # Append the predicted value to the forecasted values list
    forecasted_values.append(float(y_pred[0, 0]))

    # Update last_known_inventory for the next iteration
    last_known_inventory = y_pred

# Print the forecasted values
print("Forecasted inventory values for the next 7 days:")
for day, value in enumerate(forecasted_values, 1):
    print(f"Day {day}: {value}")

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import shapiro, kstest, jarque_bera, boxcox
from sklearn.preprocessing import PowerTransformer, StandardScaler



# Extract the column as a numpy array
data = df_new['inventory'].values

# Log Transformation
log_transformed = np.log(data + 1)

# Square Root Transformation
sqrt_transformed = np.sqrt(data)

# Box-Cox Transformation (only for positive data)
boxcox_transformed, _ = boxcox(data + 1)

# Yeo-Johnson Transformation
pt = PowerTransformer(method='yeo-johnson')
yeo_johnson_transformed = pt.fit_transform(data.reshape(-1, 1)).flatten()

# Standardization (Z-Score Normalization)
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data.reshape(-1, 1)).flatten()

# Function to check normality
def check_normality(data, title):
    plt.hist(data, bins=30)
    plt.title(f'Histogram of {title}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

    sm.qqplot(data, line='s')
    plt.title(f'Q-Q Plot of {title}')
    plt.show()

    shapiro_test = shapiro(data)
    print(f'Shapiro-Wilk Test for {title}: Statistic={shapiro_test[0]}, p-value={shapiro_test[1]}')

    ks_test = kstest(data, 'norm')
    print(f'Kolmogorov-Smirnov Test for {title}: Statistic={ks_test[0]}, p-value={ks_test[1]}')

    jb_test = jarque_bera(data)
    print(f'Jarque-Bera Test for {title}: Statistic={jb_test[0]}, p-value={jb_test[1]}')
    print()

# Check normality for each transformation
check_normality(data, 'Original Data')
check_normality(log_transformed, 'Log Transformed Data')
check_normality(sqrt_transformed, 'Square Root Transformed Data')
check_normality(boxcox_transformed, 'Box-Cox Transformed Data')
check_normality(yeo_johnson_transformed, 'Yeo-John')
check_normality(standardized_data, 'Standardized Data')
```
